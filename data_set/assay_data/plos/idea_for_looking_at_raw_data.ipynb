```python
import pandas 
import matplotlib.pyplot as plt 
plt.style.use( 'bmh' ) 
```

```python
raw = pandas.read_csv( 'plos_raw_data.csv' ) 
```

```python
summary = pandas.DataFrame( columns=[ 'sequence_position' ] ) 
```

```python
for nm, df in raw.groupby( 'sample' ):
    summary.loc[ nm ] = int( nm[1:-1] ) 
```

```python
g = summary.groupby( 'sequence_position' )
```

```python
len(g)
```

```python
fig, ax = plt.subplots( nrows=len(g), ncols=6, figsize=(10,len(g)), sharex=True ) 
for i, ( seq_pos, series ) in enumerate( g ):
    ax[ i, 0 ].set_ylabel( int(seq_pos) ) 
    for j, nm in enumerate( series.index ):
        dat = raw[(raw['sample']==nm)]
        ax[ i, j ].scatter( dat.s, dat.kobs, color='k' ) 
```
