```python
import pandas 
import numpy as np 
```

```python
tm = pandas.read_csv( 'clean_df.csv', index_col=0 ) 
# dummy cols to make merging these easier later 
tm[ 'ki' ] = np.nan
tm[ 'err_ki' ] = np.nan
plos = pandas.read_csv( 'plos/clean_plos_data_set.csv', index_col=0 ) 
```

```python
plos.head()
```

```python
thermal_kinetics = tm[ plos.columns ] 
thermal = tm[[ 'tm', 'k', 'err_tm', 'err_k' ]] 
```

```python
kinetics = pandas.concat([ thermal_kinetics[(thermal_kinetics.kcatkm>0)], plos ]) 
data_set = thermal.join( kinetics ) 
```
